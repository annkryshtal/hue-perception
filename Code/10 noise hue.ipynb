```python
def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
```

```python
import xlrd
import numpy as np
import matplotlib.pyplot as plt
file_location = "hue.xlsx"

workbook = xlrd.open_workbook(file_location)

first_sheet = workbook.sheet_by_index(1)

# print(first_sheet)
a = [360]*11
x = [first_sheet.cell_value(i, 2) for i in range(193,383,1)] + a
y = [first_sheet.cell_value(i, 3) for i in range(193,383,1)] + [first_sheet.cell_value(i, 3) for i in range(288,299,1)]

yerr_pos = [first_sheet.cell_value(i, 2) for i in range(first_sheet.ncols)]
yerr_neg = [first_sheet.cell_value(i, 3) for i in range(first_sheet.ncols)]

print(x[95], x[106])
plt.axis([-5,365,-25,25])
o = np.linspace(-25,25)
o1 = zerolistmaker(len(o))
o2 = [45]*len(o)
o3 = [90]*len(o)
o4 = [135]*len(o)
o5 = [180]*len(o)
o6 = [225]*len(o)
o7 = [270]*len(o)
o8 = [315]*len(o)
o9 = [360]*len(o)
g= np.linspace(-5,365)
h=zerolistmaker(len(g))
plt.plot(g, h, color = 'gray')
plt.plot(o1,o, color = 'gray')
plt.plot(o2,o, color = 'gray')
plt.plot(o3,o, color = 'gray')
plt.plot(o4,o, color = 'gray')
plt.plot(o5,o, color = 'gray')
plt.plot(o6,o, color = 'gray')
plt.plot(o7,o, color = 'gray')
plt.plot(o8,o, color = 'gray')
plt.plot(o9,o, color = 'gray')
plt.plot(x,y,'o',label='akryshtal')
plt.ylabel('delta, °')
plt.xlabel('angles, °')
plt.title('10 noise')
plt.legend()
plt.savefig('ak_10.png')

```
