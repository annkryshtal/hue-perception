```python
def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
```

```python
import xlrd
import numpy as np
import matplotlib.pyplot as plt
file_location = "hueak.xlsx"

workbook = xlrd.open_workbook(file_location)

first_sheet = workbook.sheet_by_index(0)

# print(first_sheet)

x = [first_sheet.cell_value(i, 5) for i in range(192,208,1)]
y = [first_sheet.cell_value(i, 7) for i in range(192,208,1)]



# print(x, y)
plt.axis([-5,360,0,10])
# plt.plot(x,y,'o')
o = np.linspace(-25,25)
o1 = zerolistmaker(len(o))
o2 = [45]*len(o)
o3 = [90]*len(o)
o4 = [135]*len(o)
o5 = [180]*len(o)
o6 = [225]*len(o)
o7 = [270]*len(o)
o8 = [315]*len(o)
o9 = [360]*len(o)
g= np.linspace(-5,365)
h=zerolistmaker(len(g))
plt.plot(g, h, color = 'gray')
plt.plot(o1,o, color = 'gray')
plt.plot(o2,o, color = 'gray')
plt.plot(o3,o, color = 'gray')
plt.plot(o4,o, color = 'gray')
plt.plot(o5,o, color = 'gray')
plt.plot(o6,o, color = 'gray')
plt.plot(o7,o, color = 'gray')
plt.plot(o8,o, color = 'gray')
plt.plot(o9,o, color = 'gray')
plt.plot(x, y, 'o', label='akryshtal')
plt.ylabel('SD, °')
plt.xlabel('angles, °')
plt.title('10 noise average ak')
plt.legend()
plt.savefig('10_ak.png')
plt.show()
```
